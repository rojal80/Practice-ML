```python
#import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2 Setup plot
fig, ax =plt.subplots(figsize=(5,8)) #(width,height)

#3.Plot data
ax.plot(x,y)

#4. customize plot
ax.set(title="Simple Plot",
       xlabel="X-axis",
       ylabel="Y-axis")

#5. Save and shor the whole figure
fig.savefig("images/sample-plot.png")
```

```python
import numpy as np
```

```python
#choose some data
x= np.linspace(0,10,100)
x[:10]
```

```python
#plot the data nad create a line plot
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

```python
#plot the line and create a scatter plot

fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#Another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from dictionary
```

```python
nut_butter_prices={"Almond butter":10,
                   "peanut butter":13,
                   "cashew butter": 15}
```

```python
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Rojal Store", ylabel="Price($)");
```
